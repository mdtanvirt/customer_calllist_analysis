```python
df_cleaned = df.apply(pd.to_numeric, errors='coerce')
df_cleaned
```

```python
numiric_df = df_cleaned.select_dtypes(include=['number'])
numiric_df
```

```python
correlation_matrix = numiric_df.corr()
correlation_matrix = correlation_matrix.dropna(how='all', axis='index').dropna(how='all', axis='columns')

correlation_matrix
```

```python
sns.heatmap(correlation_matrix.corr(), annot=True)
plt.show()
```

```python
population_columns = ['1970 Population', '1980 Population', '1990 Population', 
                      '2000 Population', '2010 Population', '2015 Population', 
                      '2020 Population', '2022 Population', 
                      'Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage']
average_population_by_continent = df.groupby('Continent')[population_columns].mean().sort_values(by='2022 Population', ascending=False)
average_population_by_continent
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
year_population_columns = ['1970 Population', '1980 Population', '1990 Population', 
                      '2000 Population', '2010 Population', '2015 Population', 
                      '2020 Population', '2022 Population']
year_average_population_by_continent = df.groupby('Continent')[year_population_columns].mean().sort_values(by='2022 Population', ascending=False)
year_average_population_by_continent

year_average_population_by_continent.plot(figsize=(16, 8))
```

```python
# transpose the data frame
transpose_df = year_average_population_by_continent.transpose()
transpose_df
```

```python
transpose_df.plot(figsize=(16, 8))
```

```python
year_average_population_by_continent.boxplot(figsize=(16, 8))
```
